```python
DATA_DIR = Path('..', 'data').resolve()
OUTPUT_DIR = DATA_DIR / 'outputs'
DATA_DIR
```

```python
plate = Plate96(max_vol=45000, dead_vol=2000)
print(plate.test())
```

```python
for ini, row in zip([f'I{x}' for x in range(1,9)] * 2, plate.rows()):
    plate.fill_block((row,), plate.columns()[2:22], ini, 2000)
print(plate)
```

```python
plate_letters_384 = 'ABCDEFGHIJKLMNOP'
m_wells = []
for i in range(0,8):
    i_helper = (i // 4) * 8  # used to assign column letters. Will be 0 for i<3 and 8 for 3<=i<6
    target_wells = [f'{plate_letters_384[k]}{j}' for k in range(i_helper, i_helper + 8)
                    for j in range(((i % 4) * 5) + 3, ((i % 4) * 5) + 8)]
    m_wells.append(target_wells)
print(m_wells)
```

```python
for mon, well_list in zip([f'M{x}' for x in range(1,9)], m_wells):
    for well in well_list:
        plate.fill_well(well, mon, 2000)
print(plate)
```

```python
plate_letters_384 = 'ABCDEFGHIJKLMNOP'
t_wells = []
for i in range(0,5):
    target_wells = [f'{plate_letters_384[k]}{j}' for j in [i + 3, i + 8, i + 13, i + 18] for k in range(0, 16)]
    t_wells.append(target_wells)
print(t_wells)
```

```python
for ter, well_list in zip([f'T{x}' for x in range(1,6)], t_wells):
    for well in well_list:
        plate.fill_well(well, ter, 2000)
print(plate)
```
